```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import DataFrame,Series
```

```python
#dataset ='pgp2'
dataset ='facebook414simi'
data = []
file_name="log/"+dataset+".txt"
for line in open(file_name,"r"): #设置文件对象并读取每一行文件
    data.append(line[:-1])               #将每一行文件加入到list中

#create_df
pgp2=DataFrame(columns=('Algorithm','Graph','Graph_class','seed_type','sample_rate','iter',
                        'pre_pivot','pre_star','pre_tie','pre_rim',
                        's_pivot','s_pivot_new','s_star','s_star_new','s_tie','s_tie_new','s_rim','s_rim_new',
                        'pivot_recall','star_recall','tie_recall','rim_recall','mean_recall',
                        'pivot_FAC','star_FAC','tie_FAC','rim_FAC','mean_FAC',
                        'pivot_MAP','star_MAP','tie_MAP','rim_MAP',
                        'pivot_NDCG','star_NDCG','tie_NDCG','rim_NDCG',
                        'KSD','KSD_pva','SDD','ND','CC','JaccardIndex'))

#add rows
for i in range(len(data)):
    if (i+1)%51 == 6:
        iters = int(data[i][-1])
        Graph = data[i][10:-2]
        Graph_class = "unbalance"
    elif (i+1)%51 == 7:
        Algorithm = data[i][15:]
    elif (i+1)%51 == 8:
        sample_rate= float(data[i][16:-2])/100
    elif (i+1)%51 == 9:
        seed_type= data[i][11:]
    elif (i+1)%51 == 17:
        pre_pivot=int(data[i][8:])
    elif (i+1)%51 == 18:
        pre_star=int(data[i][6:])
    elif (i+1)%51 == 19:
        pre_tie=int(data[i][11:])
    elif (i+1)%51 == 20:
        pre_rim=int(data[i][11:])     
    elif (i+1)%51 == 28:
        s_pivot=int(data[i][8:])
    elif (i+1)%51 == 29:
        s_pivot_new=int(data[i][12:])
    elif (i+1)%51 == 30:
        s_star=int(data[i][6:])
    elif (i+1)%51 == 31:
        s_star_new=int(data[i][10:])
    elif (i+1)%51 == 32:
        s_tie=int(data[i][11:])
    elif(i+1)%51 == 33:
        s_tie_new=int(data[i][15:])
    elif (i+1)%51 == 34:
        s_rim=int(data[i][11:])
    elif (i+1)%51 == 35:
        s_rim_new=int(data[i][15:])
    
    elif (i+1)%51 == 38:
        recall= data[i][9:-1]
    elif (i+1)%51 == 39:
        mean_recall=float(data[i][13:])#/sample_rate
    elif (i+1)%51 == 40:
        FAC=data[i][6:-1]
    elif (i+1)%51 == 41:
        mean_FAC=float(data[i][10:])
    elif (i+1)%51 == 42:
        MAP=data[i][6:-1]    
    elif (i+1)%51 == 43:
        NDCG=data[i][7:-1]
    
    elif (i+1)%51 == 45:
        #KSD = data[i][30:-1]
        #KSD_list = KSD.split(', pvalue=')
        #KSD_list=[c.strip() for c in KSD_list]
        #KSD_list=[c.replace("'-'",'nan') for c in KSD_list]
        #KSD_sta = float(KSD_list[0])
        #KSD_pva = float(KSD_list[1])
        KSD = data[i][4:]
        KSD_list = KSD.split(' ')
        KSD_sta = float(KSD_list[0])
        KSD_pva = float(KSD_list[1])
    elif (i+1)%51 == 46:
        SDD = float(data[i][5:])
    elif (i+1)%51 == 47:
        ND = float(data[i][3:])     
    elif (i+1)%51 == 48:
        CC = float(data[i][3:])
    
    elif (i+1)%51 == 49:
        JaccardIndex=float(data[i][14:])
        recall_list=recall.split(',')
        recall_list=[c.strip() for c in recall_list]
        recall_list=[c.replace("'-'",'nan') for c in recall_list]
        
        FAC_list=FAC.split(',')
        FAC_list=[c.strip() for c in FAC_list]
        FAC_list=[c.replace("'-'",'nan') for c in FAC_list]
        
        MAP_list=MAP.split(',')
        MAP_list=[c.strip() for c in MAP_list]
        MAP_list=[c.replace("'-'",'nan') for c in MAP_list]
        
        NDCG_list=NDCG.split(',')
        NDCG_list=[c.strip() for c in NDCG_list]
        NDCG_list=[c.replace("'-'",'nan') for c in NDCG_list]
        
        pivot_recall=float(recall_list[0])#/sample_rate
        star_recall=float(recall_list[1])#/sample_rate
        tie_recall=float(recall_list[2])#/sample_rate
        rim_recall=float(recall_list[3])#/sample_rate
        
        pivot_FAC=float(FAC_list[0])
        star_FAC=float(FAC_list[1])
        tie_FAC=float(FAC_list[2])
        rim_FAC=float(FAC_list[3])
        
        pivot_MAP=float(MAP_list[0])
        star_MAP=float(MAP_list[1])
        tie_MAP=float(MAP_list[2])
        rim_MAP=float(MAP_list[3])
        
        pivot_NDCG=float(NDCG_list[0])
        star_NDCG=float(NDCG_list[1])
        tie_NDCG=float(NDCG_list[2])
        rim_NDCG=float(NDCG_list[3])
        
        pgp2=pgp2.append(DataFrame({'Algorithm':[Algorithm],
                              'Graph':[Graph],
                              'Graph_class':[Graph_class],
                              'seed_type':[seed_type],
                              'sample_rate':[sample_rate],      
                              'iter':[iters],
                              'pre_pivot':[pre_pivot],
                              'pre_star':[pre_star],
                              'pre_tie':[pre_tie],
                              'pre_rim':[pre_rim],
                              's_pivot':[s_pivot],
                              's_pivot_new':[s_pivot_new],
                              's_star':[s_star],
                              's_star_new':[s_star_new],
                              's_tie':[s_tie],
                              's_tie_new':[s_tie_new],
                              's_rim':[s_rim],
                              's_rim_new':[s_rim_new],
                                  
                              'pivot_recall':[pivot_recall],
                              'star_recall':[star_recall],
                              'tie_recall':[tie_recall],
                              'rim_recall':[rim_recall],
                              'mean_recall':[mean_recall],
                                  
                              'pivot_FAC':[pivot_FAC],
                              'star_FAC':[star_FAC],
                              'tie_FAC':[tie_FAC],
                              'rim_FAC':[rim_FAC],
                              'mean_FAC':[mean_FAC],
                                  
                              'pivot_MAP':[pivot_MAP],
                              'star_MAP':[star_MAP],
                              'tie_MAP':[tie_MAP],
                              'rim_MAP':[rim_MAP],
                                  
                              'pivot_NDCG':[pivot_NDCG],
                              'star_NDCG':[star_NDCG],
                              'tie_NDCG':[tie_NDCG],
                              'rim_NDCG':[rim_NDCG],
                                    
                              'KSD':[KSD_sta],
                              'KSD_pva':[KSD_pva],
                              'SDD':[SDD],
                              'ND':[ND],
                              'CC':[CC],
                              'JaccardIndex':[JaccardIndex]}),ignore_index=True)   
print(pgp2)
```

```python
#在原有Excel中添加新数据集数据
from openpyxl import load_workbook

book = load_workbook('data_seed.xlsx')
writer = pd.ExcelWriter('data_seed.xlsx', engine='openpyxl')
writer.book = book
pgp2.to_excel(writer, 'facebook414simi')
writer.save()
```

```python
#聚合迭代次数
pgp2_mean_iter = pgp2.groupby(['Graph_class','Graph','Algorithm','seed_type','sample_rate'])['pivot_recall',
                        'star_recall','tie_recall','rim_recall','mean_recall',
                        'pivot_FAC','star_FAC','tie_FAC','rim_FAC','mean_FAC',
                        'pivot_MAP','star_MAP','tie_MAP','rim_MAP',
                        'pivot_NDCG','star_NDCG','tie_NDCG','rim_NDCG',
                        'KSD','KSD_pva','SDD','ND','CC','JaccardIndex'].mean()
pgp2_std_iter = pgp2.groupby(['Graph_class','Graph','Algorithm','seed_type','sample_rate'])['pivot_recall',
                        'star_recall','tie_recall','rim_recall','mean_recall',
                        'pivot_FAC','star_FAC','tie_FAC','rim_FAC','mean_FAC'].std()
pgp2_summary_iter=pd.merge(pgp2_mean_iter, pgp2_std_iter, on=['Graph_class','Graph','Algorithm','seed_type','sample_rate'])
print(pgp2_summary_iter)
```
